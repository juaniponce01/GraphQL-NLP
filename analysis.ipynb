```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# eliminar desc_ga_sku_producto_1
df.drop(columns=["desc_ga_sku_producto_1"], inplace=True)

df.info()
```

```python
df.value_counts("SASASA")
```

```python
df.drop(columns=["SASASA"], inplace=True)

df.info()
```

```python
# renombrar columnas

#columnas a renombrar
column_rename = {
    "id_tie_fecha_valor": "fecha",
    "id_cli_cliente": "id_cliente",
    "id_ga_vista": "id_vista",
    "id_ga_tipo_dispositivo": "id_tipo_dispositivo",
    "id_ga_fuente_medio": "id_fuente_medio",
    "desc_ga_sku_producto": "sku_producto",
    "fc_agregado_carrito_cant": "agregado_carrito_cant",
    "fc_ingreso_producto_monto": "ingreso_producto_monto",
    "fc_detalle_producto_cant": "detalle_producto_cant",
    "fc_producto_cant": "producto_cant",
    "flag_pipol": "flag_pipol",
    "id_ga_producto": "id_producto",
    "desc_ga_nombre_producto_1": "nombre_producto",
    "desc_ga_marca_producto": "marca_producto",
    "desc_ga_cod_producto": "cod_producto",
    "desc_categoria_producto": "categoria_producto",
    "desc_categoria_prod_principal": "categoria_prod_principal",
}

df.rename(columns=column_rename, inplace=True)

df.info()
```

```python
# eliminate columns with all missing values
df.dropna(axis=1, how='all', inplace=True)
df.info()
```

```python
# cambiar a formato datetime la fecha
df['fecha'] = pd.to_datetime(df['fecha'], format='%Y%m%d')

df.info()
```

```python
df.head()
```

```python
# Identificar las columnas de tipo float64
float64_columns = df.select_dtypes(include=['float64']).columns

# Convertir todas las columnas de float64 a float32
df[float64_columns] = df[float64_columns].astype('float32')

# Identificar las columnas de tipo int64
int64_columns = df.select_dtypes(include=['int64']).columns

# Convertir todas las columnas de int64 a int32
df[int64_columns] = df[int64_columns].astype('int32')

df.info()
```

```python
df = df.applymap(lambda x: None if pd.isna(x) else x)

df.head()
```
